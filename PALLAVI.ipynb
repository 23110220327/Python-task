```python
# Task 01 - Q1
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

num = 5
print("Factorial of", num, "is:", factorial(num))
```

```python
# Task 1 - Q2
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

num = 29
print(num, "is prime?", is_prime(num))
```

```python
# Task 1 - Q3
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b

terms = 10
print("Fibonacci series:")
fibonacci(terms)
```

```python
# Task 1 - Q4
def count_vowels(s):
    vowels = 'aeiouAEIOU'
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

string = "Hello World"
print("Number of vowels:", count_vowels(string))
```

```python
# Task 1 - Q5
def reverse_string(s):
    return s[::-1]

string = "Python"
print("Original:", string)
print("Reversed:", reverse_string(string))
```

```python
# Task 1 - Q6
squares = {x: x**2 for x in range(1, 11)}
print("Squares Dictionary:", squares)
```

```python
# Task 1 - Q7
text = input("Enter a string: ")
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Task 1 - Q8
num = 12345
sum_digits = sum(int(digit) for digit in str(num))
print("Sum of digits of", num, "is", sum_digits)
```

```python
# Task 3 - Sample DataFrame
import pandas as pd

data = {
    'Name': ['Amit', 'Pallavi', 'Ravi', 'Sneha', 'Meera'],
    'Subject': ['Math', 'Science', 'English', 'Math', 'English'],
    'Marks': [85, 92, 78, 89, 75]
}

df = pd.DataFrame(data)
print("Student Data:")
print(df)
```

```python
# Task 3 - Q1
print("First 5 Rows of Data:")
print(df.head())
```

```python
# Task 3 - Q2
print("Summary Statistics:")
print(df.describe())
```

```python
# Task 3 - Q3
print("Subject Frequency:")
print(df['Subject'].value_counts())
```

```python
# Task 3 - Q4
high_scores = df[df['Marks'] > 80]
print("Students with Marks > 80:")
print(high_scores)
```

```python
# Task 3 - Q5
avg_subject = df.groupby('Subject')['Marks'].mean()
print("Average Marks by Subject:")
print(avg_subject)
```

```python
# Task 3 - Q6 (Optional)
import matplotlib.pyplot as plt

plt.bar(df['Name'], df['Marks'], color='teal')
plt.title('Marks by Student')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

TASK 4:matplotlib

```python
# Task 4 - Q1
import matplotlib.pyplot as plt

x = list(range(-10, 11))
y = [i**2 for i in x]

plt.plot(x, y, label='y = x', color='blue', marker='o')
plt.title('Line Plot: y = x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Task 4 - Q2
products = ['A', 'B', 'C', 'D']
sales = [100, 150, 90, 200]

plt.bar(products, sales, color='orange')
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.grid(axis='y')
plt.show()
```

```python
# Task 4 - Q3
subjects = ['Math', 'Science', 'English', 'Art']
votes = [30, 25, 20, 15]

plt.pie(votes, labels=subjects, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.title('Subject Preferences')
plt.axis('equal')
plt.show()
```

```python
# Task 4 - Q4
x = list(range(1, 11))
y1 = [i for i in x]
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

plt.plot(x, y1, label='y = x', marker='o')
plt.plot(x, y2, label='y = x', marker='s')
plt.plot(x, y3, label='y = x', marker='^')

plt.title('Multiple Line Plots')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
